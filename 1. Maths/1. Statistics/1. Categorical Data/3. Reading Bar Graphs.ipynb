# Bar Graphs

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
players = ['John', 'Jane', 'Greg', 'Sam']
round_1_scores = [85, 90, 75, 90]
round_2_scores = [80, 95, 70, 100]

# Bar positions
x = np.arange(len(players))  # Label locations
width = 0.35  # Width of bars

# Plot
fig, ax = plt.subplots(figsize=(8, 5))
bar1 = ax.bar(x - width/2, round_1_scores, width, label='Round 1')
bar2 = ax.bar(x + width/2, round_2_scores, width, label='Round 2')

# Add labels, title, and legend
ax.set_xlabel('Players')
ax.set_ylabel('Scores')
ax.set_title('Player Scores in 2 Rounds of Pacman')
ax.set_xticks(x)
ax.set_xticklabels(players)
ax.legend()

ax.yaxis.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.7)
ax.set_yticks(range(0, 110, 10))

plt.tight_layout()
plt.show()
```

Given the bar plot above, which player's score improved the most between round 1 and 2?

**Answer**: We can see only 2 players had an improved score: *Jane* and *Sam*.

Jane's score improved by 5 points, whereas Sam's score improved by 10. Therefore, the answer is Sam.
